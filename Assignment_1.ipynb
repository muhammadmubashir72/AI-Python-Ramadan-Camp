**Question 7**

Write a Python program to input marks of 5 subjects, calculate total marks, percentage and grade.

```python
print("Enter marks of 5 subjects")

subject_1 = float(input("Subject 1: "))
subject_2 = float(input("Subject 2: "))
subject_3 = float(input("Subject 3: "))
subject_4 = float(input("Subject 4: "))
subject_5 = float(input("Subject 5: "))

total = subject_1 + subject_2 + subject_3 + subject_4 + subject_5
percentage = (total / 500) * 100

print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 80:
    grade = "A+"
elif percentage >= 70:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 50:
    grade = "C"
else:
    grade = "Fail"

print("Grade:", grade)
```

**Question 9**

Write a Python program to create a simple calculator that performs addition, subtraction, multiplication, and division based on user input.

```python
first_number = int(input("Enter First Number: "))
second_number = int(input("Enter Second Number: "))

# Operation menu
operation = input(
    "Select the operation\n"
    "1. Addition (+)\n"
    "2. Subtraction (-)\n"
    "3. Multiplication (*)\n"
    "4. Division (/)\n"
    "5. Modulus (%)\n"
    "Enter the selected value: "
)

if operation == "1":
    result = first_number + second_number
    print("Addition of Two Numbers:", result)

elif operation == "2":
    result = first_number - second_number
    print("Subtraction of Two Numbers:", result)

elif operation == "3":
    result = first_number * second_number
    print("Multiplication of Two Numbers:", result)

elif operation == "4":
    if second_number != 0:
        result = first_number / second_number
        print("Division of Two Numbers:", result)
    else:
        print("Error: Division by zero is not allowed")

elif operation == "5":
    if second_number != 0:
        result = first_number % second_number
        print("Modulus of Two Numbers:", result)
    else:
        print("Error: Division by zero is not allowed")

else:
    print("Choose the correct value")
```
