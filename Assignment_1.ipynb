**Question 7**

Write a Python program to input marks of 5 subjects, calculate total marks, percentage and grade.

```python
print("Enter marks of 5 subjects")

subject_1 = float(input("Subject 1: "))
subject_2 = float(input("Subject 2: "))
subject_3 = float(input("Subject 3: "))
subject_4 = float(input("Subject 4: "))
subject_5 = float(input("Subject 5: "))

total = subject_1 + subject_2 + subject_3 + subject_4 + subject_5
percentage = (total / 500) * 100

print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 80:
    grade = "A+"
elif percentage >= 70:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 50:
    grade = "C"
else:
    grade = "Fail"

print("Grade:", grade)
```
